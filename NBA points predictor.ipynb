```python
from IPython.display import display, Math, Latex
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime, date
import random
import os
```

```python
#player stats from 1991 to 2021 
players_1991_2021 = pd.read_csv("players.csv")
#deleat the unamed: 0 column
del players_1991_2021['Unnamed: 0']
players_1991_2021
```

```python
#2022 NBA player stats
players_2022 = pd.read_csv("nba_2022 - Sheet1 (2).csv")
#rename position column 
players_2022 = players_2022.rename(columns={"Pos\nâ–²": "Pos"})
#add year column 
players_2022["Year"] = 2022
players_2022
```

```python
#combining players stats data from 1991 to 2022 
players_df = pd.concat([players_1991_2021, players_2022])
players_df = players_df.reset_index(drop=True)
players_df
```

```python
#cleans data for players who played on more than one team in a single season 
for name, group in players_df.groupby(['Player', 'Year']):
    if len(group) > 1:
        # player played for more than one team in this year
        tot_row = group[group['Tm'] == 'TOT']
        if tot_row.empty:
            # tot_row is empty, skip this group
            continue
        tot_index = tot_row.index[0]
        # only drop rows if they exist in the dataframe
        if tot_index in players_df.index:
            # check if indexes to drop are in the dataframe
            drop_indexes = group.index[group.index != tot_index]
            if all(index in players_df.index for index in drop_indexes):
                players_df.drop(drop_indexes, inplace=True)
        # change TOT to player's last team
        last_row = group.iloc[-1]
        players_df.loc[tot_index, 'Tm'] = last_row['Tm']
        
players_df
```

```python
#codes team with a number 
teams_players_df['Team Code'] = teams_players_df["Team"].astype('category').cat.codes
```

```python
#codes player with a number 
teams_players_df['Player Code'] = teams_players_df["Player"].astype('category').cat.codes
```

```python
#remove players that have only one season of data 
teams_players_df = teams_players_df.groupby('Player Code').filter(lambda x: x.shape[0]>1)
```

```python
#n_season takes in a player and adds a columns of their next points  
def n_season(player):
    player = player.sort_values("Year")
    player["Next Pts"] = player["PTS"].shift(-1)
    return player
#applies n_season to teams_players_df
teams_players_df = teams_players_df.groupby("Player Code", group_keys = False).apply(n_season)
teams_players_df[['Player', 'Year', 'PTS', 'Next Pts']]
```

```python
#copies dataframe with all years 
teams_players_df_full = teams_players_df.copy()
```

```python
#drops na rows 
teams_players_df = teams_players_df.dropna().copy()
teams_players_df
```

```python
from sklearn.linear_model import Ridge
from sklearn.feature_selection import SequentialFeatureSelector
from sklearn.model_selection import TimeSeriesSplit, cross_val_score
```

```python
rr = Ridge(alpha=1)
split = TimeSeriesSplit(n_splits =3)
sfs = SequentialFeatureSelector(rr, n_features_to_select=30, direction="forward", cv=split, n_jobs=4)
```

```python
removed_cols = ["Next Pts", "Player", 'Tm', 'Team', 'Player Code', 'Team Code', 'Year', 'Pos']
#select all columns except for the removed ones 
selected_cols = teams_players_df.columns[~teams_players_df.columns.isin(removed_cols)]
selected_cols
```

```python
#scale each column 
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
teams_players_df.loc[:, selected_cols] = scaler.fit_transform(teams_players_df[selected_cols])
teams_players_df
```

```python
#fits the model 
sfs.fit(teams_players_df[selected_cols],teams_players_df["Next Pts"])
```

```python
#list of predictors 
predictors = list(selected_cols[sfs.get_support()])
predictors
```

```python
#backtest takes in a data, model, and predictors to train and test the data
def backtest(data, model, predictors, start=5, step=1):
    #prediction for an single year 
    all_pred = []
    years = sorted(data["Year"].unique())
    #starts from the fifth year and increases by one each year each time
    for i in range(start, len(years), step):
        current_year = years[i]
        #trains all data up to the current year 
        train = data[data["Year"] < current_year]
        #test the daat for that year
        test = data[data["Year"] == current_year]
        #fits the model 
        model.fit(train[predictors], train["Next Pts"])
        #makes the prediction
        preds = model.predict(test[predictors])
        #creates a df with columns acutal and prediction 
        preds = pd.Series(preds, index=test.index)
        combined = pd.concat([test["Next Pts"], preds],axis=1)
        combined.columns = ["actual", "prediction"]
        
        all_pred.append(combined)
    return pd.concat(all_pred, axis=0)
```

```python
#calls backtest function 
predictions = backtest(teams_players_df, rr,predictors)
predictions
```

```python
from sklearn.metrics import mean_squared_error
#calculates the mean sqares error
mean_squared_error(predictions["actual"], predictions["prediction"])
```

```python
def player_hist(df):
    df = df.sort_values("Year")
    df["player_year"] = range(0, df.shape[0])
    
    # Find the player's peak
    peak_year = df['PTS'].idxmax()
    peak_value = df.loc[peak_year, 'PTS']
    
    # Calculate the correlation between player_year and Pts, using the peak as a reference
    df["Pts corr"] = list(df[["player_year" ,'PTS']].expanding().corr().loc[(slice(None), 'player_year'), 'PTS'])
    df["Pts corr"].fillna(1, inplace=True)
    
    # Adjust the correlation value by how close it is to the player's peak and which side of the peak the player is on
    df["Pts corr"] *= 1 - abs(peak_year - df['player_year']) / df.shape[0]
    df["Pts corr"] *= np.sign(peak_value - df['PTS'])
    
    df["Pts diff"] = df['PTS'] / df["PTS"].shift(1)
    df["Pts diff"].fillna(1, inplace=True)
    df["Pts diff"][df["Pts diff"] == np.inf] = 1
    
    return df
teams_players_df = teams_players_df.groupby('Player Code', group_keys=False).apply(player_hist)


'''This adjusts the correlation value by both how close the player is to their peak and which side of the peak they are on. If the player is on the same side of the peak as their current performance (i.e. above the peak if they are currently performing well, or below the peak if they are currently performing poorly), the correlation value will be higher. Conversely, if the player is on the opposite side of the peak as their current performance (i.e. below the peak if they are currently performing well, or above the peak if they are currently performing poorly), the correlation value will be lower. This should give you a better idea of how well a player is performing relative to their peak and their current position relative to their peak, which should help make more accurate predictions about future points.'''
```

```python
#compares players points to the average points score in that year  
def group_avg(df):
    return df["PTS"] / df["PTS"].mean()
```

```python
teams_players_df["pts year"] = teams_players_df.groupby("Year",group_keys=False).apply(group_avg)
```
