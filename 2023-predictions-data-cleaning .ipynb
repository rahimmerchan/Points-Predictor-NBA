# Creating a 2023-2024 DF

```python
#set max columns 
pd.set_option('display.max_columns', None)
```

```python
#read in cleaned_players
new_df = pd.read_csv("cleaned_players.csv")
```

```python
new_df.head()
```

```python
# Calculate the previous year
new_df['Previous_Year'] = new_df['Year'] - 1

# Create the season column
new_df['season'] = new_df['Previous_Year'].astype(str) + '-' + new_df['Year'].astype(str)

# Drop the columns that were used for calculations (optional)
new_df.drop(['Previous_Year'], axis=1, inplace=True)

# Now the "Modified_Year" column represents the range of years (e.g., "1990-1991")
new_df.head()
```

```python
#convert Team to a coded value
new_df['Team Code'] = new_df["Team"].astype('category').cat.codes
```

```python
#convert Player to a coded value
new_df['Player Code'] = new_df["Player"].astype('category').cat.codes
```

```python
#remove players that only played one year
new_df = new_df.groupby('Player Code').filter(lambda x: x.shape[0]>1)
```

```python
new_df.head()
```

```python
#delete the unessary columns
del new_df['Unnamed: 0']
```

```python
#n_season takes in a player and adds a columns of their next points  
def n_season(player):
    player = player.sort_values("Year")
    player["Next Pts"] = player["PTS"].shift(-1)
    return player
#applies n_season to teams_players_df
new_df = new_df.groupby("Player Code", group_keys = False).apply(n_season)
```

```python
new_df.head()
```

```python
#create a list of the columns to remove
removed_cols = ["Next Pts", "Player", 'Tm', 'Team', 'Player Code', 'Team Code', 'Year', 'Pos', 'season']
#select all columns except for the removed ones 
selected_cols = new_df.columns[~new_df.columns.isin(removed_cols)]
selected_cols
```

```python
from sklearn.preprocessing import MinMaxScaler
#scale the selected columns
scaler = MinMaxScaler()
new_df.loc[:, selected_cols] = scaler.fit_transform(new_df[selected_cols])
new_df.head()
```

```python
def player_hist(df):
    df = df.sort_values("Year")
    df["player_year"] = range(0, df.shape[0])
    
    # Find the player's peak
    peak_year = df['PTS'].idxmax()
    peak_value = df.loc[peak_year, 'PTS']
    
    # Calculate the correlation between player_year and Pts, using the peak as a reference
    correlations = df[["player_year", 'PTS']].expanding().corr().loc[(slice(None), 'player_year'), 'PTS'].values
    df["Pts corr"] = correlations
    df["Pts corr"].fillna(1, inplace=True)
    
    # Adjust the correlation value by how close it is to the player's peak and which side of the peak the player is on
    proximity_to_peak = 1 - abs(peak_year - df['player_year']) / df.shape[0]
    performance_relative_to_peak = np.sign(peak_value - df['PTS'])
    adjusted_corr = correlations * proximity_to_peak * performance_relative_to_peak
    df["Adj Pts corr"] = adjusted_corr
    
    df["Pts diff"] = df['PTS'] / df["PTS"].shift(1)
    df["Pts diff"].fillna(1, inplace=True)
    df["Pts diff"][df["Pts diff"] == np.inf] = 1
    
    return df

# Assuming 'teams_players_df' is your DataFrame containing player data
new_df = new_df.groupby('Player Code', group_keys=False).apply(player_hist)
```

```python
def group_avg(df):
    return df["PTS"] / df["PTS"].mean()
new_df["pts_year"] = new_df.groupby("Year",group_keys=False).apply(group_avg)
```

```python
#Sort the dataframe by Player and Year
new_df = new_df.sort_values(by=['Player', 'Year'])

#Add a new column to flag missing previous year's data
new_df['Missing_Prev_Year'] = new_df.groupby('Player')['Year'].diff() > 1

#Convert boolean values to 1 and 0
new_df['Missing_Prev_Year'] = new_df['Missing_Prev_Year'].astype(int)

new_df.head()
```

```python
#Sort the dataframe by Player and Year
new_df = new_df.sort_values(by=['Player', 'Year'])

#Calculate the total games played by each player over all years
total_games_played = new_df.groupby('Player')['G'].sum().reset_index()
total_games_played.rename(columns={'G': 'Total_Games_Played'}, inplace=True)
new_df = pd.merge(new_df, total_games_played, on='Player', how='left')

#Calculate the average games played per year for each player
new_df['Avg_Games_Per_Year'] = new_df['Total_Games_Played'] / new_df.groupby('Player')['Year'].transform('nunique')

#Calculate the change in games played compared to the previous year
new_df['Change_Games_Played'] = new_df.groupby('Player')['G'].diff()

new_df.head()
```

```python
#Sort the dataframe by Player and Year
new_df = new_df.sort_values(by=['Player', 'Year'])

# Calculate the change in games played compared to the previous year
new_df['Change_Games_Played'] = new_df.groupby('Player')['G'].diff()

#Define a threshold to identify a significant drop in games played
significant_drop_threshold = -0.5

# Create a binary feature indicating significant drop in games played
new_df['Significant_Drop_Games_Played'] = (new_df['Change_Games_Played'] <= significant_drop_threshold).astype(int)

new_df.head()
```

```python

# alculate the overall injury metric
new_df['Overall_Injury_Metric'] = (
    new_df['Missing_Prev_Year'] * 0.2 +
    new_df['Total_Games_Played'] * 0.1 +
    new_df['Avg_Games_Per_Year'] * 0.1 +
    new_df['Change_Games_Played'] * 0.3 +
    new_df['Significant_Drop_Games_Played'] * 0.3
)

#Normalize the overall injury metric to the range [0, 1]
new_df['Normalized_Injury_Metric'] = (new_df['Overall_Injury_Metric'] - new_df['Overall_Injury_Metric'].min()) / (new_df['Overall_Injury_Metric'].max() - new_df['Overall_Injury_Metric'].min())

new_df.head()
```

```python
columns_to_fill = ['Change_Games_Played', 'Overall_Injury_Metric', 'Normalized_Injury_Metric', 'Avg_Games_Per_Year', 'Total_Games_Played', 'Adj Pts corr']

# Fill null values in the specified columns with 0
new_df.fillna({col: 0 for col in columns_to_fill}, inplace=True)
```
